```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 방 사이즈 정의
room_size = 300
fig, ax = plt.subplots(figsize=(8, 8))

# 방 전체 틀
ax.add_patch(patches.Rectangle((0, 0), room_size, room_size, fill=False, edgecolor='black', linewidth=2))

# 방향 (남쪽이 위쪽)
ax.annotate('South ↑ (Window)', xy=(280, 20), xytext=(280, -5), ha='center', fontsize=10, color='orange',
            arrowprops=dict(facecolor='orange', shrink=0.05, width=1, headwidth=6))
ax.annotate('North ↓ (Entrance / Door)', xy=(20, 280), xytext=(20, 295), ha='center', fontsize=10, color='black',
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))

# 선반
ax.add_patch(patches.Rectangle((0, 0), 20, 200, edgecolor='black', facecolor='lightgray'))
ax.text(10, 100, 'Shelf', ha='center', va='center', fontsize=8, rotation=90)

# 식탁 본체 (주황색)
ax.add_patch(patches.Rectangle((20, 0), 80, 80, edgecolor='brown', facecolor='orange'))
# 반원 (주황색)
table_arc = patches.Wedge(center=(60, 80), r=40, theta1=0, theta2=180,
                          facecolor='orange', edgecolor='brown')
ax.add_patch(table_arc)
ax.text(60, 40, 'Table Area', ha='center', va='center', fontsize=8, color='saddlebrown')

# 공기청정기 (중심 좌표 275, 95)
ax.add_patch(patches.Circle((275, 95), radius=25, edgecolor='gray', facecolor='lightgray'))
ax.text(275, 95, 'Air\nPurifier', ha='center', va='center', fontsize=7)

# 쥑쥑이 케이지
ax.add_patch(patches.Rectangle((20, 125), 52, 73, edgecolor='black', facecolor='pink', linestyle='--'))
ax.text(46, 161.5, 'Rat Cage\n(73×52×132)', ha='center', va='center', fontsize=8, color='deeppink')

# 건조기
ax.add_patch(patches.Rectangle((241, 0), 49, 59, edgecolor='green', facecolor='lightgreen'))
ax.text(265.5, 29.5, 'Dryer', ha='center', va='center', fontsize=8, color='darkgreen')
# 건조기 gap
ax.add_patch(patches.Rectangle((290, 0), 10, 59, edgecolor='none', facecolor='green', alpha=0.2))
ax.text(295, 29.5, 'Gap', ha='center', va='center', fontsize=6, color='darkgreen', rotation=90)

# 소파
ax.add_patch(patches.Rectangle((230, 125), 70, 175, edgecolor='blue', facecolor='lightblue'))
ax.text(265, 212, 'Sofa', ha='center', va='center', fontsize=8, color='navy')

# 러그
ax.add_patch(patches.Rectangle((130, 125), 100, 170, edgecolor='purple', facecolor='plum'))
ax.text(180, 210, 'Rug', ha='center', va='center', fontsize=8, color='purple')

# 부여구 길
ax.add_patch(patches.Rectangle((0, 200), 20, 100, edgecolor='none', facecolor='gray', alpha=0.3))
ax.text(10, 250, 'To\nKitchen', ha='center', va='center', fontsize=7)

# 복도
ax.add_patch(patches.Rectangle((0, 280), 70, 20, edgecolor='none', facecolor='gray', alpha=0.3))
ax.text(35, 290, 'Hallway', ha='center', va='center', fontsize=7)

# 월패드
ax.add_patch(patches.Rectangle((70, 280), 60, 20, edgecolor='none', facecolor='darkgray', alpha=0.3))
ax.text(100, 290, 'Wallpad', ha='center', va='center', fontsize=7, color='white')

# 방문
ax.add_patch(patches.Rectangle((130, 280), 170, 20, edgecolor='none', facecolor='dimgray', alpha=0.3))
ax.text(215, 290, 'Room Door', ha='center', va='center', fontsize=7, color='white')

# 창문 라벨 (ub: 50cm height)
ax.text(150, -10, 'Window (Starts at 50cm height)', ha='center', va='top', fontsize=7, color='darkorange')

# 컨센트
ax.plot(0, 25, 'ro')
ax.text(0, 25, '🔌', fontsize=10, ha='right')
ax.plot(300, 25, 'ro')
ax.text(300, 25, '🔌', fontsize=10, ha='left')
ax.plot(300, 275, 'ro')
ax.text(300, 275, '🔌', fontsize=10, ha='left')

# 시각화 설정
ax.set_xlim(0, room_size)
ax.set_ylim(0, room_size)
ax.set_aspect('equal')
ax.set_title("공기청정기 중심 (275,95) + 식탁 주흑색", fontsize=13)
ax.invert_yaxis()
plt.tight_layout()
plt.show()
```
